## Mantenedora 

```python
Mantenedora = pd.read_excel("fies1.xlsx", usecols="D, C, E")
Mantenedora.head(11)
```

```python
## Verificação de dados nulos
Mantenedora.isna().sum()
```

```python
## Verificação de dados duplicados 
Mantenedora.duplicated().sum()
```

```python
## Removendo dados duplicados
Mantenedora = Mantenedora.drop_duplicates()
Mantenedora_count = Mantenedora.count()[0] 

Mantenedora_count
```

```python
pd.DataFrame({"Total de Mantenedora" : [Mantenedora_count]})
```

```python
Mantenedora.to_csv("mantenedora.csv")
```
